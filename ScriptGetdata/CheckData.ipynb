```python
# read csv file and print out the data
import pandas as pd
import os

# read csv file
def get_year_data(file_path):
    data = pd.read_csv(file_path)
    # get unique year
    years = data["Year"].unique()
    # get missing years: 2000-2024
    missing_years = [str(year) for year in range(2000, 2025) if year not in years]

    # get type from file name between _ and .csv
    file_name = file_path.split("/")[-1]
    base_name = file_name.rsplit('_', 1)[0]
    type = file_name.split('_')[-1].split(".")[0]

    # convert to dataframe:
    row = {"Name": base_name}
    for year in years:
        row[str(year)] = type
    for year in missing_years:
        row[year] = "Missing"
    result_df = pd.DataFrame([row])

    # sort columns
    result_df = result_df[["Name"] + [str(year) for year in range(2000, 2025)]]

    return result_df

# list of csv files
file_list = os.listdir("../Data/CleanData/")
file_paths = ["../Data/CleanData/" + file for file in file_list]
```

```python
get_year_data(file_paths[0])
```

```python
# analysis all files
result = pd.DataFrame()
for file_path in file_paths:
    # row bind all data
    result = pd.concat([result, get_year_data(file_path)], ignore_index=True)

# write to csv
result.to_csv("../Data/GetData.csv", index=False)
```
